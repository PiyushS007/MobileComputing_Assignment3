```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt


df = pd.read_csv('/content/accelerometer_data_200L.csv')


train_df = df.iloc[:-10]  # Use all rows except the last 10 for training
valid_df = df.iloc[-10:]  # Last 10 rows for validation


X_train = train_df[['X', 'Y', 'Z']]
y_train = train_df[['X', 'Y', 'Z']]
X_valid = valid_df[['X', 'Y', 'Z']]

model = RandomForestRegressor()
model.fit(X_train, y_train)


predictions = model.predict(X_valid)


plt.figure(figsize=(15, 10))


time_labels = np.arange(0, 2, 0.2)


plt.subplot(3, 1, 1)
plt.plot(time_labels, valid_df['X'], label='Actual X', marker='o')
plt.plot(time_labels, predictions[:, 0], label='Predicted X', marker='x')
plt.xlabel('Time (seconds)')
plt.ylabel('Value')
plt.title('Predicted vs Actual Values for X Axis')
plt.legend()


plt.subplot(3, 1, 2)
plt.plot(time_labels, valid_df['Y'], label='Actual Y', marker='o')
plt.plot(time_labels, predictions[:, 1], label='Predicted Y', marker='x')
plt.xlabel('Time (seconds)')
plt.ylabel('Value')
plt.title('Predicted vs Actual Values for Y Axis')
plt.legend()


plt.subplot(3, 1, 3)
plt.plot(time_labels, valid_df['Z'], label='Actual Z', marker='o')
plt.plot(time_labels, predictions[:, 2], label='Predicted Z', marker='x')
plt.xlabel('Time (seconds)')
plt.ylabel('Value')
plt.title('Predicted vs Actual Values for Z Axis')
plt.legend()

plt.tight_layout()
plt.show()
```
